```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

# Load the dataset
df = pd.read_csv('cleaned_students_performance.csv')

# Display the first few rows
print("First 5 rows of the dataset:")
print(df.head())

# Display data types and non-null values
print("\nDataFrame Info:")
print(df.info())

# Display descriptive statistics
print("\nDescriptive Statistics:")
print(df.describe())

# Check for missing values
print("\nMissing values per column:")
print(df.isnull().sum())

# Check for duplicate values
print("\nNumber of duplicate rows:")
print(df.duplicated().sum())
```

* The dataset consists of 986 entries and 10 columns.
* There are no missing values or duplicate rows, indicating a clean dataset for analysis.
* Data types are appropriate, with 3 numerical columns (math_score, reading_score, writing_score) and 7 categorical columns (gender, race/ethnicity, parental_level_of_education, lunch, test_preparation_course, family_support, alcohol_consumption).
*************
* Descriptive Statistics:
* Scores (Math, Reading, Writing):
* math_score: Mean = 66.69, Median = 66.5, Std = 14.34
* reading_score: Mean = 69.72, Median = 70.0, Std = 13.91
* writing_score: Mean = 68.65, Median = 69.0, Std = 14.43
* All scores have a similar range (27âˆ’100), and their means and medians are relatively close, suggesting a somewhat symmetrical distribution.

```python
# Set style for plots
sns.set_style("whitegrid")

# 1. Distributions of Numerical Features (Histograms)
plt.figure(figsize=(18, 5))

plt.subplot(1, 3, 1)
sns.histplot(df['math_score'], kde=True, bins=20, color='skyblue')
plt.title('Distribution of Math Score')
plt.xlabel('Math Score')
plt.ylabel('Frequency')

plt.subplot(1, 3, 2)
sns.histplot(df['reading_score'], kde=True, bins=20, color='lightcoral')
plt.title('Distribution of Reading Score')
plt.xlabel('Reading Score')
plt.ylabel('Frequency')

plt.subplot(1, 3, 3)
sns.histplot(df['writing_score'], kde=True, bins=20, color='lightgreen')
plt.title('Distribution of Writing Score')
plt.xlabel('Writing Score')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()

# 2. Counts of Categorical Features (Bar Plots)
categorical_cols = ['gender', 'race/ethnicity', 'parental_level_of_education', 'lunch', 'test_preparation_course', 'family_support', 'alcohol_consumption']

plt.figure(figsize=(20, 15))
for i, col in enumerate(categorical_cols):
    plt.subplot(3, 3, i + 1)
    sns.countplot(y=df[col], palette='viridis', order=df[col].value_counts().index)
    plt.title(f'Distribution of {col.replace("_", " ").title()}')
    plt.xlabel('Count')
    plt.ylabel('')
plt.tight_layout()
plt.show()
```

*Distributions of Test Scores (Histograms):

 * The histograms for math_score, reading_score, and writing_score show roughly normal (bell-shaped) distributions, with a slight tendency for more students to score in the higher ranges (60âˆ’80).
 * This indicates that while most students are performing around the average, there are also students achieving very high scores.

*Distributions of Categorical Features (Bar Plots):

* Gender: Slightly more female students (51.5%) than male students (48.5%).
* Race/Ethnicity: Group C is the most represented group (32%), followed by Group D (26%), Group B (19%), Group E (13%), and Group A (10%).
* Parental Level of Education: 'Some college' and 'associate's degree' are the most common education levels for parents, suggesting a significant portion of students come from families with mid-level educational backgrounds.
* Lunch: The majority of students (65%) have a 'standard' lunch, while a considerable number (35%) have 'free/reduced' lunch.
* Test Preparation Course: More students (65%) did 'not complete' the test preparation course compared to those who 'completed' it (35%).
* Family Support: Almost half of the students (49.9%) have family support, and the other half (50.1%) do not.
* Alcohol Consumption: More students (59.6%) consume alcohol than those who do not (40.4%).

```python
# 3. Relationships between Scores and Categorical Features (Box Plots)
plt.figure(figsize=(20, 15))

# Math Score vs. Gender
plt.subplot(2, 3, 1)
sns.boxplot(x='gender', y='math_score', data=df, palette='pastel')
plt.title('Math Score by Gender')
plt.xlabel('Gender')
plt.ylabel('Math Score')

# Reading Score vs. Parental Level of Education
plt.subplot(2, 3, 2)
sns.boxplot(x='parental_level_of_education', y='reading_score', data=df, palette='pastel',
            order=['some high school', 'high school', 'some college', "associate's degree", "bachelor's degree", "master's degree"])
plt.title('Reading Score by Parental Level of Education')
plt.xlabel('Parental Level of Education')
plt.ylabel('Reading Score')
plt.xticks(rotation=45, ha='right')

# Writing Score vs. Test Preparation Course
plt.subplot(2, 3, 3)
sns.boxplot(x='test_preparation_course', y='writing_score', data=df, palette='pastel')
plt.title('Writing Score by Test Preparation Course')
plt.xlabel('Test Preparation Course')
plt.ylabel('Writing Score')

# Math Score vs. Lunch
plt.subplot(2, 3, 4)
sns.boxplot(x='lunch', y='math_score', data=df, palette='pastel')
plt.title('Math Score by Lunch')
plt.xlabel('Lunch')
plt.ylabel('Math Score')

# Reading Score vs. Family Support
plt.subplot(2, 3, 5)
sns.boxplot(x='family_support', y='reading_score', data=df, palette='pastel')
plt.title('Reading Score by Family Support')
plt.xlabel('Family Support')
plt.ylabel('Reading Score')

# Writing Score vs. Alcohol Consumption
plt.subplot(2, 3, 6)
sns.boxplot(x='alcohol_consumption', y='writing_score', data=df, palette='pastel')
plt.title('Writing Score by Alcohol Consumption')
plt.xlabel('Alcohol Consumption')
plt.ylabel('Writing Score')

plt.tight_layout()
plt.show()

# 4. Relationships between Numerical Features (Scatter Plots)
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
sns.scatterplot(x='math_score', y='reading_score', data=df, color='blue', alpha=0.6)
plt.title('Math Score vs. Reading Score')
plt.xlabel('Math Score')
plt.ylabel('Reading Score')

plt.subplot(1, 3, 2)
sns.scatterplot(x='math_score', y='writing_score', data=df, color='green', alpha=0.6)
plt.title('Math Score vs. Writing Score')
plt.xlabel('Math Score')
plt.ylabel('Writing Score')

plt.subplot(1, 3, 3)
sns.scatterplot(x='reading_score', y='writing_score', data=df, color='red', alpha=0.6)
plt.title('Reading Score vs. Writing Score')
plt.xlabel('Reading Score')
plt.ylabel('Writing Score')

plt.tight_layout()
plt.show()
```

Relationships between Scores and Categorical Features (Box Plots):

* Math Score by Gender: Male students generally show higher math scores than female students, although the difference is not extremely large.
* Reading Score by Parental Level of Education: There's a clear positive trend: students whose parents have higher levels of education (e.g., master's degree, bachelor's degree) tend to achieve higher reading scores. This suggests parental education level is a significant factor in reading performance.
* Writing Score by Test Preparation Course: Students who completed the test preparation course show noticeably higher writing scores compared to those who not completed it. This indicates the effectiveness of the test preparation course.
* Math Score by Lunch: Students with 'standard' lunch tend to have higher math scores than those with 'free/reduced' lunch. This could be indicative of socioeconomic factors impacting performance.
* Reading Score by Family Support: Students with 'yes' for family support tend to have slightly higher reading scores than those with 'no' family support.
* Writing Score by Alcohol Consumption: Students with 'no' alcohol consumption tend to have slightly higher writing scores than those with 'yes' alcohol consumption.

Relationships between Numerical Features (Scatter Plots):
* The scatter plots between any two of the three scores (math_score vs. reading_score, math_score vs. writing_score, reading_score vs. writing_score) show a strong positive linear relationship. This means students who perform well in one subject are highly likely to perform well in the other two, suggesting a general academic aptitude across subjects.

```python
# 5. Correlation Analysis
numerical_cols = ['math_score', 'reading_score', 'writing_score']
correlation_matrix = df[numerical_cols].corr()

print("\nCorrelation Matrix:")
print(correlation_matrix)

plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation Matrix of Test Scores')
plt.show()

# 6. Mean and Median Calculation
print("\nMean of Scores:")
print(df[numerical_cols].mean())

print("\nMedian of Scores:")
print(df[numerical_cols].median())

# Mean and Median by Gender
print("\nMean Scores by Gender:")
print(df.groupby('gender')[numerical_cols].mean())

print("\nMedian Scores by Gender:")
print(df.groupby('gender')[numerical_cols].median())

# Mean and Median by Test Preparation Course
print("\nMean Scores by Test Preparation Course:")
print(df.groupby('test_preparation_course')[numerical_cols].mean())

print("\nMedian Scores by Test Preparation Course:")
print(df.groupby('test_preparation_course')[numerical_cols].median())

# Mean and Median by Parental Level of Education
print("\nMean Scores by Parental Level of Education:")
print(df.groupby('parental_level_of_education')[numerical_cols].mean().sort_values(by='math_score'))

print("\nMedian Scores by Parental Level of Education:")
print(df.groupby('parental_level_of_education')[numerical_cols].median().sort_values(by='math_score'))

# Mean and Median by Lunch
print("\nMean Scores by Lunch:")
print(df.groupby('lunch')[numerical_cols].mean())

print("\nMedian Scores by Lunch:")
print(df.groupby('lunch')[numerical_cols].median())
```
